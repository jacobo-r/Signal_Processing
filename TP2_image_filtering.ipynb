```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Sobel filter
The Sobel filter is a gradient-based filter used for edge detection. It computes the gradient of the image intensity at each pixel, giving the direction of the largest intensity change.

- `dx`: Represents the horizontal (x-direction) gradient of the image. It emphasizes vertical edges.
- `dy`: Represents the vertical (y-direction) gradient of the image. It emphasizes horizontal edges.

### Gradient filter
Computes the magnitude of the gradient, giving an overall measure of edge strength. Gradient Magnitude Formula:
$$
\text{gradient\_magnitude} = \sqrt{{\text{sobel\_x}}^2 + {\text{sobel\_y}}^2}
$$

### Averaging filter
The averaging filter (mean filter) smoothens the image by replacing each pixel's value with the average value of its neighbors. The size (5,5) indicates that it considers a 5x5 neighborhood around each pixel. This filter is useful for reducing noise but can also blur edges.

### Gaussian filter
The Gaussian filter is another smoothing filter using a Gaussian kernel. It gives more weight to the nearby pixels and less weight to the distant pixels. This results in a smoother and less abrupt blurring effect compared to the `Averaging filter`.

```python
# Load the image
image = cv2.imread('barbara.jpeg', cv2.IMREAD_GRAYSCALE)

# Sobel Filters
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)

# Gradient Filter
gradient_magnitude = cv2.magnitude(sobelx, sobely)

# Averaging Filter
averaging_result = cv2.blur(image, (5,5))

# Gaussian Filter
gaussian_result = cv2.GaussianBlur(image,(5,5),cv2.BORDER_DEFAULT)
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Display original image
axs[0, 0].imshow(image, cmap='gray')
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

# Display Sobel X Filter
axs[0, 1].imshow(sobelx, cmap='gray')
axs[0, 1].set_title('Sobel X')
axs[0, 1].axis('off')

# Display Sobel Y Filter
axs[0, 2].imshow(sobely, cmap='gray')
axs[0, 2].set_title('Sobel Y')
axs[0, 2].axis('off')

# Display Gradient Filter
axs[1, 0].imshow(gradient_magnitude, cmap='gray')
axs[1, 0].set_title('Gradient Magnitude')
axs[1, 0].axis('off')

# Display Averaging Filter
axs[1, 1].imshow(averaging_result, cmap='gray')
axs[1, 1].set_title('Averaging Filter')
axs[1, 1].axis('off')

# Display Gaussian Filter
axs[1, 2].imshow(gaussian_result, cmap='gray')
axs[1, 2].set_title('Gaussian Filter')
axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
```
